```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'Arial'
from matplotlib import colors
import spe2py as spe
from cellpose import models
import cellpose
import time
```

```python
spe_tool = spe.load()
```

```python
frame_count = 0
for i in spe_tool.file.data[:]:
    frame_count += 1
print(frame_count)

roi_count = 0
for i in spe_tool.file.data[0]:
    roi_count += 1
print(roi_count)
```

```python
# 7 frames, 1 ROI
frame_data = spe_tool.file.data[0][0] # [frame][ROI]
#frame_data.shape = (640, 512)
```

```python
for i in range(0, 7):
    fig = plt.figure(figsize=(5, 5))
    ax = fig.add_axes([0.2, 0.2, 0.7, 0.7])
    mappable = ax.imshow(spe_tool.file.data[i][0], cmap='hot')
    fig.colorbar(mappable ,fraction=0.0495, pad=0.04) #set the colorbar size match to the contour size
```

```python
temp = []
for i in range(0, 7):
    temp.append(spe_tool.file.data[i][0][0][10])

coef = np.polyfit(np.arange(0, 7), temp, 1)
ploy1d_f = np.poly1d(coef)

plt.plot(np.arange(0, 7), temp, 'ro')
plt.plot(np.arange(0, 7), ploy1d_f(np.arange(0, 7)), 'b--')
```

```python
# Fit intensity at different pixles in 7 different time frames
x = np.arange(0, 7)
slope_matrix = np.zeros((640, 512))

t0 = time.perf_counter()
for i in range(0, 512):
    for j in range(0, 640):
        intensity_frame = []
        for k in range(0, 7):
            intensity_frame.append(spe_tool.file.data[k][0][j][i])
        coef = np.polyfit(x, intensity_frame, 1)
        slope_matrix[j][i] = coef[0]

t1 = time.perf_counter()
print(t1-t0, 'seconds')

'''
# Test if slope matrix is correct
temp = []
for i in range(0, 7):
    temp.append(spe_tool.file.data[i][0][0][10])

np.polyfit(np.arange(0, 7), temp, 1)[0] == slope_matrix[0][10]
'''
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_axes([0.2, 0.2, 0.7, 0.7])

mappable = ax.imshow(slope_matrix, cmap='gray')

fig.colorbar(mappable, fraction=0.0495, pad=0.04) #set the colorbar size match to the contour size
```

```python
t0 = time.perf_counter()

# model_type='cyto' or 'nuclei' or 'cyto2'
model = models.Cellpose(model_type='cyto')

# list of files
files = ['no1_1.tiff']

imgs = [cellpose.io.imread(f) for f in files]
nimg = len(imgs)

# define CHANNELS to run segementation on
# grayscale=0, R=1, G=2, B=3
# channels = [cytoplasm, nucleus]
# if NUCLEUS channel does not exist, set the second channel to 0
channels = [[0,0]]
# IF ALL YOUR IMAGES ARE THE SAME TYPE, you can give a list with 2 elements
# channels = [0,0] # IF YOU HAVE GRAYSCALE
# channels = [2,3] # IF YOU HAVE G=cytoplasm and B=nucleus
# channels = [2,1] # IF YOU HAVE G=cytoplasm and R=nucleus

# if diameter is set to None, the size of the cells is estimated on a per image basis
# you can set the average cell `diameter` in pixels yourself (recommended)
# diameter can be a list or a single number for all images

masks, flows, styles, diams = model.eval(imgs, diameter=None, channels=channels)

t1 = time.perf_counter()
```

```python
cellpose.io.masks_flows_to_seg(images=imgs,
                               masks=masks,
                               flows=flows,
                               diams=diams,
                               file_names=['no1_1'],
                               channels=None)
```
